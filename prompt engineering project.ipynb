# How many people have French "fr" as their Language?

# Job Title Contains Engineer 


```python
# Assuming your DataFrame is named 'df'
import re  # Import the regular expression library

# Use a regular expression with the 're' library to find job titles containing "Engineer" (case-insensitive)
# The 're.I' flag makes the search case-insensitive
engineer_jobs = df[df['Job'].str.contains(r'engineer', flags=re.I, regex=True)]

# Get the number of engineers by counting the rows in the filtered DataFrame
num_engineers = len(engineer_jobs)

# Print the number of engineers
print(f"Number of job titles containing 'Engineer' (case-insensitive): {num_engineers}")
```

# Find the Email of the person with following IP Address :149.146.147.205

```python
# Assuming your DataFrame is named 'df'

# Specify the IP address you want to search for
target_ip = "149.146.147.205"

# Use boolean indexing to filter rows where 'IP Address' matches the target IP
filtered_rows = df[df['IP Address'] == target_ip]

# Check if any rows match the target IP
if not filtered_rows.empty:
    # Extract the email address from the first matching row (assuming unique email addresses per IP)
    target_email = filtered_rows.iloc[0]['Email']
    print(f"Email of the person with IP Address {target_ip}: {target_email}")
else:
    print(f"No person found with IP Address {target_ip}")
```

# How many People have Mastercard as their Credit Card Provider and made a purchase above $50

```python
# Assuming your DataFrame is named 'df'

# Filter the rows where 'CC Provider' is 'Mastercard' and 'Purchase Price' is above $50
mastercard_above_50 = df[(df['CC Provider'] == 'Mastercard') & (df['Purchase Price'] > 50)]

# Get the number of people who meet both criteria
num_mastercard_above_50 = len(mastercard_above_50)

# Print the result
print(f"Number of people with Mastercard and a purchase above $50: {num_mastercard_above_50}")
```

# Find the email of the person with following Credit card number : 561252141909

```python
# Assuming your DataFrame is named 'df'

# Specify the credit card number you want to search for
target_credit_card = 561252141909

# Use boolean indexing to filter rows where 'Credit Card' matches the target credit card number
filtered_rows = df[df['Credit Card'] == target_credit_card]

# Check if any rows match the target credit card number
if not filtered_rows.empty:
    # Extract the email address from the first matching row (assuming unique email addresses per credit card)
    target_email = filtered_rows.iloc[0]['Email']
    print(f"Email of the person with Credit Card Number {target_credit_card}: {target_email}")
else:
    print(f"No person found with Credit Card Number {target_credit_card}")
```

# How many people purchase during the AM and how many people purchase during PM 

```python
# Assuming your DataFrame is named 'df'

# Use boolean indexing to filter rows where 'AM or PM' is 'AM'
am_purchases = df[df['AM or PM'] == 'AM']

# Use boolean indexing to filter rows where 'AM or PM' is 'PM'
pm_purchases = df[df['AM or PM'] == 'PM']

# Get the counts of purchases in AM and PM
num_am_purchases = len(am_purchases)
num_pm_purchases = len(pm_purchases)

# Print the results
print(f"Number of people who made purchases during AM: {num_am_purchases}")
print(f"Number of people who made purchases during PM: {num_pm_purchases}")
```

# How many people have a credit that expires in 2020?

```python
# Assuming your DataFrame is named 'df'

# Extract the year from the 'CC Exp Date' column using string manipulation
# We assume that the 'CC Exp Date' column contains dates in the format 'MM/YY'
# We split the date on '/' and extract the second part (YY), then convert it to an integer
df['CC Exp Year'] = df['CC Exp Date'].apply(lambda x: int(x.split('/')[1]))

# Use boolean indexing to filter rows where the credit card expires in 2020 (CC Exp Year is 20)
expires_in_2020 = df[df['CC Exp Year'] == 20]

# Get the count of people whose credit card expires in 2020
num_people_expiring_in_2020 = len(expires_in_2020)

# Print the result
print(f"Number of people with a credit card expiring in 2020: {num_people_expiring_in_2020}")
```

# what are the top 5 most popular email providers

```python
# Assuming your DataFrame is named 'df'

# Extract the email domain from the 'Email' column using string manipulation
df['Email Domain'] = df['Email'].apply(lambda x: x.split('@')[1])

# Count the occurrences of each email domain and get the top 5
top_5_email_providers = df['Email Domain'].value_counts().head(5)

# Print the result
print("Top 5 most popular email providers:")
print(top_5_email_providers)
```

## Distribution of Credit Card Expiration Years

```python
import matplotlib.pyplot as plt

# Assuming you've already extracted 'expires_in_2020' DataFrame from a previous answer

# Count the number of people with credit cards expiring in each year
expiring_years_count = expires_in_2020['CC Exp Year'].value_counts()

# Create a bar chart to visualize the distribution of credit card expiration years
plt.figure(figsize=(8, 6))
plt.bar(expiring_years_count.index, expiring_years_count.values)
plt.xlabel('Expiration Year')
plt.ylabel('Number of People')
plt.title('Credit Card Expiration Years in 2020')
plt.show()
```

##  Purchase Distribution in AM and PM

```python
import matplotlib.pyplot as plt

# Assuming you've already calculated 'num_am_purchases' and 'num_pm_purchases' from a previous answer

# Data for the pie chart
labels = ['AM Purchases', 'PM Purchases']
sizes = [num_am_purchases, num_pm_purchases]
colors = ['pink', 'green']

# Create a pie chart to visualize the distribution of purchases in AM and PM
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Purchase Distribution in AM and PM')
plt.show()
```
